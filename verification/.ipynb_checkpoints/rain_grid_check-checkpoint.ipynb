```python
import os
from datetime import datetime, timedelta
from glob import glob
from multiprocessing import Pool

import numpy as np
import xarray as xr
from scipy.ndimage import map_coordinates

from matplotlib import pyplot as plt
from tqdm import tqdm

import pyproj
```

```python
def explore_rrate(rrate_rf_ffn, rrate_dp_ffn, dp_scaling=1):

    #open data
    with xr.open_dataset(rrate_rf_ffn) as ds:
        rrate_rf = ds['rain_rate'].data
        x = ds.x.data
        y = ds.y.data
        x_grid, y_grid = np.meshgrid(x, y)
        grid_res = (x[1]-x[0]) * 1000 #m
    with xr.open_dataset(rrate_dp_ffn) as ds:
        rrate_dp = ds['rain_rate'].data * dp_scaling
        
    #plot
    plt.figure(figsize=[14,12])
    
    ax = plt.subplot(221)
    plt.title('RF3')
    c = plt.pcolor(x_grid, y_grid, rrate_rf, vmin=0, vmax=80)
    plt.colorbar(c, ax=ax, label='rain rate (mmh-1)')
    
    ax = plt.subplot(222)
    plt.title('DP')
    c = plt.pcolor(x_grid, y_grid, rrate_dp, vmin=0, vmax=80)
    plt.colorbar(c, ax=ax, label='rain rate (mmh-1)')
    
    plt.subplot(223)
    x_val = rrate_rf.flatten()
    y_val = rrate_dp.flatten()
    filt = np.logical_and(x_val>0, y_val>0)
    x_val = x_val[filt]
    y_val = y_val[filt]
    plt.plot(x_val, y_val, 'k.', ms=0.5)
    plt.xlim([0, 80])
    plt.ylim([0, 80])
    plt.xlabel('RF3 rain rate (mmh-1)')
    plt.ylabel('DP rain rate (mmh-1)')
    plt.grid()
    plt.plot([0,80],[0,80],'r:')
    m, b = np.polyfit(x_val, y_val, 1)
    x_fit = np.array([0,80])
    plt.plot(x_fit, m*x_fit + b, 'b-')
    print(m)
    
    plt.subplot(224)
    plt.hist([rrate_rf.flatten(), rrate_dp.flatten()], np.arange(7.5,182.5,5), label=['RF3', 'DP'], histtype='step', linewidth=2, density=True)
    plt.legend(loc='upper right')
    plt.xlim([0, 80])
    plt.xlabel('rain rate (mmh-1)')
    plt.ylabel('Normalised Frequency')
```

```python
# #Melbourne 30 Jan 2019 case
# #storms are in the NE quadrant, moving SE
rrate_rf_ffn = '/g/data/rq0/admin/dp-rain-verification/02_rrate_orig/2019/01/30/2_20190130_050045.prcp-rrate.nc'
rrate_dp_ffn = '/g/data/rq0/admin/dp-rain-verification/02_rrate_dp/2019/01/30/2_20190130_050045.prcp-rrate.nc'
explore_rrate(rrate_rf_ffn, rrate_dp_ffn, dp_scaling=1)
```
