```python
#30 Jan 2019 case
#storms are in the NE quadrant, moving SE

optflow_1_ffn = '/g/data/rq0/admin/dp-rain-verification/02_optflow/2019/01/30/2_20190130_045442.optflow.nc'
optflow_2_ffn = '/g/data/rq0/admin/dp-rain-verification/02_optflow/2019/01/30/2_20190130_050045.optflow.nc'
rrate_1_ffn = '/g/data/rq0/admin/dp-rain-verification/02_rrate_dp/2019/01/30/2_20190130_045442.prcp-rrate.nc'
rrate_2_ffn = '/g/data/rq0/admin/dp-rain-verification/02_rrate_dp/2019/01/30/2_20190130_050045.prcp-rrate.nc'
```

```python
#open data
with xr.open_dataset(optflow_1_ffn) as ds:
    oflow1_u = ds['flow_u'].data
    oflow1_v = ds['flow_v'].data #units of m/s
    x = ds.x.data
    y = ds.y.data
    x_grid, y_grid = np.meshgrid(x, y)
    grid_res = (x[1]-x[0]) * 1000 #m
    print(grid_res)
with xr.open_dataset(optflow_2_ffn) as ds:
    oflow2_u = ds['flow_u'].data
    oflow2_v = ds['flow_v'].data
with xr.open_dataset(rrate_1_ffn) as ds:
    rrate1 = ds['rain_rate'].data
    #rrate1[rrate1==0] = np.nan
with xr.open_dataset(rrate_2_ffn) as ds:
    rrate2 = ds['rain_rate'].data
    #rrate2[rrate2==0] = np.nan

td = 6*60 #timestep in seconds
```

```python
#convert oflow from m/s to pix/min
oflow1_u_pix = oflow1_u*60/grid_res #units of pix/min
oflow1_v_pix = oflow1_v*60/grid_res
oflow2_u_pix = oflow2_u*60/grid_res
oflow2_v_pix = oflow2_v*60/grid_res
```

```python
def advection(rrate1, rrate2,
              oflow1_u_pix, oflow1_v_pix, oflow2_u_pix, oflow2_v_pix,
              T_start=0, T_end=6,
              T=6, t=1):
    """
    WHAT:
    Returns the accumulated rainfall amount by applying a time weighted interpolation from radar volume 1 to radar volume 2
    using optical flow fields (u,v and both timesteps) and an interpolation timestep of t. Can return r_acc only for a portion of the timespan
    between radar volumes 1 and 2.
    
    HELP
    note: first radar volume is the oldest. second radar volume is the newest.

    INPUTS
    rrate1/rrate2: rain rate (mm/hr) for first radar volume and second radar volume
    oflow1_u_pix/oflow1_v_pix/oflow2_u_pix/oflow2_v_pix: optical flow (pix/min) in u and v direction for first radar volume and second radar volume
    T_start: starting timestep (minutes from radar volume 1) for interpolation (0 will include the first radar timestep)
    T_end: ending timestep (minutes from radar volume 2) for interpolation (T_end=T will include the second radar timestep)
    T: time difference between radar volumes (minutes)
    t: timestep for interpolation (minutes)
    
    OUTPUTS:
    r_acc: accumulated rainfall totals (mm)
    """

    # Perform temporal interpolation
    r_acc = np.zeros((rrate1.shape))
    x, y = np.meshgrid(
        np.arange(rrate1.shape[1], dtype=float), np.arange(rrate1.shape[0], dtype=float)
    )
    for i in range(T_start, T_end + t, t):
        #shift timestep 1 forwards (this is the older timestep)
        ts_forward = -i
        y1_shift = y + (ts_forward * oflow1_v_pix)
        x1_shift = x + (ts_forward * oflow1_u_pix)
        pos1 = (y1_shift, x1_shift)
        R1 = map_coordinates(rrate1, pos1, order=1)
        weight1 = (T - i)/T

        #shift timestep 2 backwards (this is the newer timestep)
        ts_backwards = (T-i)
        y2_shift = y + (ts_backwards * oflow2_v_pix)
        x2_shift = x + (ts_backwards * oflow2_u_pix)
        pos2 = (y2_shift, x2_shift)
        R2 = map_coordinates(rrate2, pos2, order=1)
        weight2 = i/T

        #weighted combination
        rrate_interp = R1 * weight1 + R2 * weight2

        #convert to mm/hr in mm
        r_acc += rrate_interp/60*t
        
    return r_acc
        
#         #plotting
#         xlim = [-60,-30]
#         ylim = [80,110]

#         plt.figure(figsize=[10,10])
#         skip_step = 5
#         skip = (slice(None,None,skip_step),slice(None,None,skip_step))

#         plt.subplot(221)
#         plt.title('Timestep 1 - original')
#         plt.pcolor(x_grid, y_grid, rrate1, vmin=0, vmax=100, cmap='jet')
#         plt.quiver(x_grid[skip], y_grid[skip], oflow1_u_pix[skip], -oflow1_v_pix[skip], scale=10, scale_units='inches')
#         plt.xlim(xlim)
#         plt.ylim(ylim)
#         plt.grid()

#         plt.subplot(222)
#         plt.title(f'Timestep 1 - advection at {i} with weight {weight1}')
#         plt.pcolor(x_grid, y_grid, R1, vmin=0, vmax=100, cmap='jet')
#         plt.xlim(xlim)
#         plt.ylim(ylim)
#         plt.grid()

#         plt.subplot(223)
#         plt.title('Timestep 2 - original')
#         plt.pcolor(x_grid, y_grid, rrate2, vmin=0, vmax=100, cmap='jet')
#         plt.quiver(x_grid[skip], y_grid[skip], oflow2_u_pix[skip], -oflow2_v_pix[skip], scale=10, scale_units='inches')
#         plt.xlim(xlim)
#         plt.ylim(ylim)
#         plt.grid()

#         plt.subplot(224)
#         plt.title(f'Timestep 2 - advection at {i} with weight {weight2}')
#         plt.pcolor(x_grid, y_grid, R2, vmin=0, vmax=100, cmap='jet')
#         plt.xlim(xlim)
#         plt.ylim(ylim)
#         plt.grid()


    #     fig = plt.figure(figsize=[10,5])
    #     ax = plt.subplot(121)
    #     c = plt.pcolor(x_grid, y_grid, rrate_interp, vmin=0, vmax=100, cmap='hot_r')
    #     fig.colorbar(c, ax=ax)
    #     plt.xlim(xlim)
    #     plt.ylim(ylim)
    #     plt.grid()

    #     ax = plt.subplot(122)
    #     c = plt.pcolor(x_grid, y_grid, r_acc, vmin=0, vmax=6, cmap='hot_r')
    #     fig.colorbar(c, ax=ax)
    #     plt.xlim(xlim)
    #     plt.ylim(ylim)
    #     plt.grid()
```

```python
r_acc = advection(rrate1, rrate2, oflow1_u_pix, oflow1_v_pix, oflow2_u_pix, oflow2_v_pix, T_start=0, T_end=6, T=6, t=1)
    
#plotting
xlim = [-100,-0]
ylim = [0,100]

fig = plt.figure(figsize=[18,5])
skip_step = 5
skip = (slice(None,None,skip_step),slice(None,None,skip_step))

ax = plt.subplot(131)
plt.title('Timestep t - original')
c = plt.pcolor(x_grid, y_grid, rrate1, vmin=0, vmax=100, cmap='jet')
fig.colorbar(c, ax=ax, label='mm/hr')
plt.quiver(x_grid[skip], y_grid[skip], oflow1_u_pix[skip], -oflow1_v_pix[skip], scale=10, scale_units='inches')
plt.xlim(xlim)
plt.ylim(ylim)
plt.grid()

ax = plt.subplot(132)
plt.title('Timestep t+1 - original')
c = plt.pcolor(x_grid, y_grid, rrate2, vmin=0, vmax=100, cmap='jet')
fig.colorbar(c, ax=ax, label='mm/hr')
plt.quiver(x_grid[skip], y_grid[skip], oflow2_u_pix[skip], -oflow2_v_pix[skip], scale=10, scale_units='inches')
plt.xlim(xlim)
plt.ylim(ylim)
plt.grid()

ax = plt.subplot(133)
plt.title('Accumulation using advection')
c = plt.pcolor(x_grid, y_grid, r_acc, vmin=0, vmax=6, cmap='jet')
fig.colorbar(c, ax=ax, label='mm')
plt.xlim(xlim)
plt.ylim(ylim)
plt.grid()
```
