```python
import pyart
import matplotlib.pyplot as pl
import warnings
from matplotlib import pyplot as plt
from matplotlib.colors import LogNorm
from csu_radartools import csu_kdp
from scipy import integrate
import numpy as np
from numba import jit
```

```python
warnings.simplefilter("ignore")
```

```python
@jit
def fill_phi(phi):
    nx, ny = phi.shape
    for i in range(nx):
        phi_val = 0
        for j in range(ny):
            if np.isnan(phi[i, j]):
                phi[i, j] = phi_val
            else:
                phi_val = phi[i, j]
                
    return phi

def phidp_bringi(radar, gatefilter, phidp_field="PHI_UNF", refl_field='DBZ'):
    """
    Compute PHIDP and KDP Bringi.
    Parameters
    ==========
    radar:
        Py-ART radar data structure.
    gatefilter:
        Gate filter.
    unfold_phidp_name: str
        Differential phase key name.
    refl_field: str
        Reflectivity key name.
    Returns:
    ========
    phidpb: ndarray
        Bringi differential phase array.
    kdpb: ndarray
        Bringi specific differential phase array.
    """
    dz = radar.fields[refl_field]['data'].copy().filled(-9999)
    dp = radar.fields[phidp_field]['data'].copy().filled(-9999)

    # Extract dimensions
    rng = radar.range['data']
    azi = radar.azimuth['data']
    dgate = rng[1] - rng[0]
    [R, A] = np.meshgrid(rng, azi)

    # Compute KDP bringi.
    kdpb, phidpb, _ = csu_kdp.calc_kdp_bringi(dp, dz, R / 1e3, gs=dgate, bad=-9999, thsd=12, window=6.0, std_gate=11)

    # Mask array
    phidpb = np.ma.masked_where(phidpb == -9999, phidpb)
    kdpb = np.ma.masked_where(kdpb == -9999, kdpb)
    
    #fill
    phidpb = fill_phi(phidpb.filled(np.NaN))
    
    # Get metadata.
    phimeta = pyart.config.get_metadata("differential_phase")
    phimeta['data'] = phidpb
    kdpmeta = pyart.config.get_metadata("specific_differential_phase")
    kdpmeta['data'] = kdpb

    return phimeta, kdpmeta

def _fix_phidp_from_kdp(phidp, kdp, r, gatefilter):
    """
    Correct PHIDP and KDP from spider webs.

    Parameters
    ==========
    r:
        Radar range.
    gatefilter:
        Gate filter.
    kdp_name: str
        Differential phase key name.
    phidp_name: str
        Differential phase key name.

    Returns:
    ========
    phidp: ndarray
        Differential phase array.
    """
    kdp[gatefilter.gate_excluded] = 0
    kdp[(kdp < -4)] = 0
    kdp[kdp > 15] = 0
    interg = integrate.cumtrapz(kdp, r, axis=1)

    phidp[:, :-1] = interg / (len(r))
    return phidp, kdp

def phidp_giangrande(radar, gatefilter, refl_field='DBZ', ncp_field='NCP',
                     rhv_field='RHOHV_CORR', phidp_field='PHIDP', VERBOSE=False):
    """
    Phase processing using the LP method in Py-ART. A LP solver is required,

    Parameters:
    ===========
    radar:
        Py-ART radar structure.
    gatefilter:
        Gate filter.
    refl_field: str
        Reflectivity field label.
    ncp_field: str
        Normalised coherent power field label.
    rhv_field: str
        Cross correlation ration field label.
    phidp_field: str
        Differential phase label.

    Returns:
    ========
    phidp_gg: dict
        Field dictionary containing processed differential phase shifts.
    kdp_gg: dict
        Field dictionary containing recalculated differential phases.
    """
#     #this function also hardcodes the gatefilter to the phase field
#     if VERBOSE:
#         print('unwrap phase')
#     unfphidic = pyart.correct.dealias_unwrap_phase(radar,
#                                                    gatefilter=gatefilter,
#                                                    skip_checks=True,
#                                                    vel_field=phidp_field,
#                                                    nyquist_vel=90)
# #     #hardcode gatefilter
# #     phidp_tmp = unfphidic['data'].copy()
# #     phidp_tmp[gatefilter.gate_excluded] = np.NaN
# #     phidp_tmp = np.ma.masked_invalid(phidp_tmp)
# #     np.ma.set_fill_value(phidp_tmp, np.NaN)
# #     unfphidic['data'] = phidp_tmp
    
#     radar.add_field_like(phidp_field, 'PHITMP', unfphidic['data'], replace_existing=True)
    if VERBOSE:
        print('lp proc phase')
    phidp_gg, kdp_gg = pyart.correct.phase_proc_lp_gf(radar, gatefilter=gatefilter,
                                                   LP_solver='cylp',
                                                   refl_field=refl_field,
                                                   phidp_field=phidp_field)    
    
#     phidp_gg, kdp_gg = pyart.correct.phase_proc_lp(radar, 0.0,
#                                                    LP_solver='cylp',
#                                                    ncp_field=ncp_field,
#                                                    refl_field=refl_field,
#                                                    rhv_field=rhv_field,
#                                                    phidp_field='PHITMP')
    if VERBOSE:
        print('fix phase from kdp')
    
    phidp_gg['data'], kdp_gg['data'] = _fix_phidp_from_kdp(phidp_gg['data'],
                                                           kdp_gg['data'],
                                                           radar.range['data'],
                                                           gatefilter)

    try:
        # Remove temp variables.
        radar.fields.pop('unfolded_differential_phase')
        radar.fields.pop('PHITMP')
    except Exception:
        pass

    phidp_gg['data'] = phidp_gg['data'].astype(np.float32)
    phidp_gg['_Least_significant_digit'] = 4
    kdp_gg['data'] = kdp_gg['data'].astype(np.float32)
    kdp_gg['_Least_significant_digit'] = 4

    return phidp_gg, kdp_gg
```

```python
def analyse_phidp(vol_ffn):
    #load data
    radar = pyart.aux_io.read_odim_h5(vol_ffn, file_field_names=True)
    gatefilter = pyart.filters.GateFilter(radar)
    
    #shift the system phase
    rawphi = radar.fields["PHIDP"]['data']
    sysphase = pyart.correct.phase_proc.det_sys_phase(radar, ncp_field="RHOHV", rhv_field="RHOHV", phidp_field="PHIDP", rhohv_lev=0.98)
    radar.add_field_like("PHIDP", "PHIDP_shifted", rawphi - sysphase, replace_existing=True)
    print('system phase:', round(sysphase))
    
    #unfold phase
    unfphidic = pyart.correct.dealias_unwrap_phase(radar,
                                               gatefilter=gatefilter,
                                               skip_checks=True,
                                               vel_field='PHIDP_shifted',
                                               nyquist_vel=180)
    radar.add_field_like("PHIDP", "PHIDP_unwrap", unfphidic['data'], replace_existing=True)
    
    
    #calculate phidp from bringi technique
    phidp_b, kdp_b = phidp_bringi(radar, gatefilter, phidp_field="PHIDP_unwrap", refl_field='DBZH')
    radar.add_field("PHIDP_B", phidp_b, replace_existing=True)
    
    #calculate phidp from LP technique
    phidp_g, kdp_g = phidp_giangrande(radar, gatefilter, refl_field='DBZH', ncp_field='RHOHV',
                    rhv_field='RHOHV', phidp_field='PHIDP_unwrap', VERBOSE=False)    
    radar.add_field("PHIDP_G", phidp_g, replace_existing=True)
    
    fig = plt.figure(figsize=[14,8])
    tilt = 1
    gr = pyart.graph.RadarDisplay(radar)
    plt.subplot(231)
    gr.plot_ppi("PHIDP_shifted", tilt, vmin=0, vmax=90, cmap="pyart_Wild25")      
    plt.subplot(232)
    gr.plot_ppi("PHIDP_unwrap", tilt, vmin=0, vmax=90, cmap="pyart_Wild25")    
    plt.subplot(233)
    gr.plot_ppi("PHIDP_B", tilt, vmin=0, vmax=90, cmap="pyart_Wild25", title='PHIDP Bringi')       
    plt.subplot(234)
    gr.plot_ppi("PHIDP_G", tilt, vmin=0, vmax=90, cmap="pyart_Wild25", title='PHIDP LP')
    
    
    
    sl = radar.get_slice(0)
    phi0 = radar.fields["PHIDP_unwrap"]["data"][sl].copy()
    phib = radar.fields["PHIDP_B"]["data"][sl].copy()
    phig = radar.fields["PHIDP_G"]["data"][sl].copy()
    phib[phi0.mask] = np.NaN
    phig[phi0.mask] = np.NaN
    
    plt.subplot(235)
    plt.hist2d(phi0.flatten(), phib.flatten(), range=[[0, 90], [0, 90]], bins=45, norm=LogNorm(1, 5e3))
    plt.ylabel("PHIDP")
    plt.xlabel("PHIDP Bringi")
    plt.colorbar()

    plt.subplot(236)
    plt.hist2d(phi0.flatten(), phig.flatten(), range=[[0, 90], [0, 90]], bins=45, norm=LogNorm(1, 5e3))
    plt.ylabel("PHIDP")
    plt.xlabel("PHIDP LP")
    plt.colorbar()
```
